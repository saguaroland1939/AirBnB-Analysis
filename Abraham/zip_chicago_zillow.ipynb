```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
engine = create_engine('postgresql://postgres:postgres@localhost:5432/zillow_airbnb_db')
```

```python
zillow_zip_df = pd.read_csv("zillow_Zip_time_series copy.csv")
zillow_zip_df
```

```python
zillow_zip_na= zillow_zip_df[['Date','RegionName','MedianListingPricePerSqft_AllHomes','MedianListingPrice_AllHomes']]

zillow_zip_na
```

```python
five_year_df = zillow_zip_na.loc[zillow_zip_na['Date'] >= '2012-1-1']
five_year_df
```

```python
chicago_region_df = five_year_df.loc[(five_year_df['RegionName'] >= 60601) & (five_year_df['RegionName'] <= 60827)]
final_chicago_df = chicago_region_df.drop_duplicates()
final_chicago_df
```

```python
final_chicago_data = final_chicago_df.dropna()
final_chicago_data
```

```python
final_zip_df = final_chicago_data.groupby('RegionName').agg({"MedianListingPricePerSqft_AllHomes": "mean", 'MedianListingPrice_AllHomes':"mean"}).reset_index()          
final_df = final_zip_df.rename(columns = {'RegionName':'zip_code', 'MedianListingPricePerSqft_AllHomes': 'avg_price_per_sq_ft', 'MedianListingPrice_AllHomes': 'avg_price'},index = {'RegionName':'zip_code'})                                                            
final_df                                                            
                                                            
```

```python
final_df.to_sql(name='chicago_zip',
                        con=engine,
                        index=False,
                        if_exists='append')
```

```python
final_df.to_csv('../zillow_zip.csv')
```
